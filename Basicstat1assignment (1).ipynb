```python
#Identifying numerical columns in the dataset.

import pandas as pd

df = pd.read_csv("sales_data_with_discounts.csv")

numerical_cols = df.select_dtypes(include=['number']).columns
numerical_cols
```

```python
# Calculating the mean, median, mode, and standard deviation for these columns.



def calculate_stats(df, column):
    mean = df[column].mean()
    median = df[column].median()
    mode = df[column].mode().iloc[0]  # Get the first mode if multiple exist
    std_dev = df[column].std()
    return mean, median, mode, std_dev

for col in numerical_cols:
    mean, median, mode, std_dev = calculate_stats(df, col)
    print(f"Statistics for column {col}:")
    print(f"  Mean: {mean}")
    print(f"  Median: {median}")
    print(f"  Mode: {mode}")
    print(f"  Standard Deviation: {std_dev}")
    print("-" * 20)
```

```python
#	Providing a brief interpretation of these statistics.

# The code calculates descriptive statistics (mean, median, mode, standard deviation) for numerical columns in a DataFrame.
#  To interpret the results:
#
# Mean: Represents the average value of the column.  Sensitive to outliers.
# Median:  The middle value when the data is sorted. Less affected by extreme values than the mean.
# Mode:  The most frequent value. Useful for categorical data, but can be less informative for continuous numerical data.
# Standard Deviation: Measures the amount of variation or dispersion of a set of values. A high standard deviation indicates more spread out data.
#
# Comparing the mean and median can give you insights into the distribution of data:
#   - If mean ≈ median:  The data is likely symmetrically distributed (e.g., normal distribution).
#   - If mean > median: The data is likely skewed to the right (positive skew), meaning there are some high values pulling the mean upwards.
#   - If mean < median: The data is likely skewed to the left (negative skew), meaning there are some low values pulling the mean downwards.

# Example Interpretation (hypothetical):
#
# Statistics for column UnitPrice:
#   Mean: 10.5
#   Median: 9.8
#   Mode: 5
#   Standard Deviation: 3.1
#
#  In this example, the mean is slightly higher than the median. This suggests a slight positive skew in the distribution of UnitPrice, meaning there are a few high-priced items. The mode of 5 suggests that many items are priced at this level. The standard deviation of 3.1 shows that the prices have moderate variability.
```

```python
# Plotting histograms for each numerical column.

import matplotlib.pyplot as plt

for col in numerical_cols:
  plt.figure(figsize=(8, 6))  # Adjust figure size as needed
  plt.hist(df[col], bins=20, edgecolor='black')  # Adjust bins as needed
  plt.xlabel(col)
  plt.ylabel('Frequency')
  plt.title(f'Histogram of {col}')
  plt.show()
```

```python
#  Analyzing the distribution (e.g., skewness, presence of outliers) and provide inferences.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()

    # Calculate skewness
    skewness = df[col].skew()
    print(f"Skewness for {col}: {skewness}")

    # Interpret skewness
    if abs(skewness) < 0.5:
        print(f"The distribution of {col} is approximately symmetric.")
    elif 0.5 <= abs(skewness) < 1:
        print(f"The distribution of {col} is moderately skewed.")
    else:
        print(f"The distribution of {col} is highly skewed.")

    # Identify outliers using IQR
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"Number of outliers in {col}: {len(outliers)}")
    if len(outliers) > 0:
      print(f"Example outlier values in {col}: {outliers[col].head()}")
    print("-" * 20)
```

```python
# Identifing categorical columns in the dataset.

categorical_cols = df.select_dtypes(exclude=['number']).columns
categorical_cols
```

```python
#Creating bar charts to visualize the frequency or count of each category.
# 	Analyze the distribution of categories and provide insights.

# Assuming 'df' and 'categorical_cols' are defined as in  provided code.

for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
    plt.tight_layout()  # Adjust layout to prevent labels from overlapping
    plt.show()

    # Analyze the distribution
    category_counts = df[col].value_counts()
    total_count = len(df)

    print(f"Analysis for {col}:")
    for category, count in category_counts.items():
        percentage = (count / total_count) * 100
        print(f"- {category}: {count} ({percentage:.2f}%)")
    print("-" * 20)
```

```python
# Explaining the concept of standardization (z-score normalization).
# 	Standardize the numerical columns using the formula: z=x-mu/sigma
# 	Show before and after comparisons of the data distributions

# Standardize numerical features
for col in numerical_cols:
    # Calculate mean and standard deviation
    mu = df[col].mean()
    sigma = df[col].std()

    # Apply standardization
    df[col + '_standardized'] = (df[col] - mu) / sigma

# Display before and after comparisons
for col in numerical_cols:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Before standardization
    sns.histplot(df[col], kde=True, ax=axes[0])
    axes[0].set_title(f'Distribution of {col} (Before Standardization)')

    # After standardization
    sns.histplot(df[col + '_standardized'], kde=True, ax=axes[1])
    axes[1].set_title(f'Distribution of {col} (After Standardization)')

    plt.show()
```

```python
#Discussing the need for converting categorical data into dummy variables (one-hot encoding).
# 	Apply one-hot encoding to the categorical columns, creating binary (0 or 1) columns for each category.
# 	Display a portion of the transformed dataset

# One-hot encode categorical columns
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Display a portion of the transformed dataset
print(df.head())
```
